```python
ki_file = path.join(doc_path,"Temblor-RMN(depurada-incluye controles)KI.xlsx")
ki_df = pd.read_excel(ki_file)
ki_df = ki_df[ki_df['registro tableta'].notna()].set_index('clinica')
filtered_ki_df = ki_df.iloc[:,:6].drop('MoCA',1)
filtered_ki_df
```

```python
doc_path = "/data/elekin/doc"
metadata_file = path.join(doc_path,"Temblor-RMN(depurada-controles-ethw)IS.xlsx")
metadf=pd.read_excel(metadata_file)
etonadf = metadf[metadf.etona]
etonadf.set_index(etonadf.columns[0]).iloc[:,:4]
```

```python
metadata = etonadf.set_index(etonadf.columns[0]).iloc[:,:4].join(filtered_ki_df)
metadata.temblor = metadata.temblor.fillna('no')
metadata.TRS = metadata.TRS.fillna(0).astype('int8')
metadata['F.Nacimiento'] = metadata['F.Nacimiento'].fillna(0).astype('int16')
metadata['age'] = 2015 - metadata['F.Nacimiento']
metadata = metadata.drop('registro tableta',1)
metadata.index.name='name'
metadata
```

```python
import datetime

today=datetime.date.today().strftime("%Y%m%d")
metadata.to_csv(path.join(doc_path,"metadata-202106_{}.csv".format(today)))
```
