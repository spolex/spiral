```python
from __future__ import absolute_import, division, print_function, unicode_literals

import h5py
import logging
from os import path
from pandas import HDFStore
import pandas as pd
import tensorflow as tf
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping

%matplotlib inline
```

```python
FEATURES = 4096
coeff = 17

root_path = "Z:/elekin"
rdo_root_path = path.join(root_path,"02-RESULTADOS/03-HANDWRITTING")
h5_outputfile = path.join(rdo_root_path, "00-OUTPUT/archimedean-")
h5_output_filename = h5_outputfile + str(coeff) + "-splits" +".h5"
hdf = HDFStore(h5_output_filename)

hdf.keys()
```

```python
X_train, y_train, X_test, y_test = hdf[hdf.keys()[0]], hdf[hdf.keys()[1]], hdf[hdf.keys()[2]], hdf[hdf.keys()[3]]
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
y_train=y_train.values.astype('int8')
X_train = X_train.values.astype('float32')
y_test=y_test.values.astype('int8')
X_test = X_test.values.astype('float32')
```

```python
#Early stop configuration
earlystop_callback = EarlyStopping(
  monitor='val_accuracy', min_delta=0.0001,
  patience=2)

training_earlystop_callback = EarlyStopping(
  monitor='accuracy', min_delta=1e-6,
  patience=10)
```

```python
tf.random.set_seed(2019) # establecemos la semilla para tensorflow
np.random.seed(2019) # establecemos la semilla para numpy
```

```python
# RNN architectures
tf.keras.backend.clear_session()# para evitar que entrenamientos annteriores afecten
model = tf.keras.models.Sequential([
tf.keras.layers.Lambda(lambda x: tf.expand_dims(x, axis=-1),# expand the dimension form (50, 4096) to (50, 4096, 1)
                      input_shape=(FEATURES,)),
tf.keras.layers.LayerNormalization(),
tf.keras.layers.LSTM(8, activation=tf.nn.tanh, return_sequences=False),
tf.keras.layers.Dropout(0.5),
tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)])
```

```python
model.summary()
```

```python
# define loss function and compile model
opt = tf.keras.optimizers.Adam(learning_rate=1e-6)
model.compile(loss="binary_crossentropy", optimizer=opt, metrics=["accuracy"])
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=200, batch_size=10, use_multiprocessing=True, callbacks=[training_earlystop_callback])
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
