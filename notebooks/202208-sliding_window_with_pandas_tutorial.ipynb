```python
import pandas as pd
from functools import reduce
import numpy as np
```

```python
labels = pd.DataFrame.from_dict({'c1': 1, 'c2': 0, 'c3':0},orient='index', columns=['labels'])
labels
```

```python
def make_sliding(df, N, s=1):
    dfs = [df.shift(-i).dropna().applymap(lambda x: [x]) for i in range(0, N + 1, s)]
    return reduce(lambda x, y: x.add(y), dfs).dropna()
```

```python
df = pd.DataFrame(dict(c1 = [0,1,2,3,4,5,6,5,5,5,56,6], c2 = [2,4,2,4,2,4,2,3,4,5,6,8], c3 = [3,5,6,3,4,6,9,8,0,7,3,4]))
df=df
df
```

```python
df1 = make_sliding(df,5,1)
df1.values.shape
```

```python
df1
```

```python
df2 = df1.T.reset_index().melt(["index"])
df2
```

```python
df3 = df2.drop("variable", axis=1).set_index("index")
df3
```

```python
df4=df3.apply(lambda x: x.value, axis=1, result_type='expand')
df4
```
